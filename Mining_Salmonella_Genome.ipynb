```python
#Open Fasta file with Salmonella Genome 

with open("salmonella_spi1_region.fna", "r") as f: 
    SALMONELLA = f.read()
```

```python
# Remove the heading on the Fasta file 

SALMONELLA_LIST = SALMONELLA.split('\n')[1:]
SALMONELLA_LIST = ''.join(SALMONELLA_LIST)
```

```python
def gc_map(seq, block_size, gc_thresh):
    
    #List to store 4 nucleotide blocks 
    BLOCKS = []
    
    #index to keep track of which block number you are in. 
    J = 0 
    
    #add in each block depending on the specified block size for the entire genome from beginning to end/
    while J < len(seq)/block_size: 
        BLOCKS.append(seq[J: J+block_size])
        J += 1
    
    #list to store the GC content of each block
    GC= []
    
    #calculate the GC content for each block
    
    for i in range(0, len(BLOCKS)): 
        GC.append((BLOCKS[i].count('G')+ BLOCKS[i].count('C'))/gc_thresh)
        
    #Return the genome of segments GC content above the pathogenicity threshold in lowercase. 
    
    for i in range(0, len(GC)): 
        if GC[i] <gc_thresh:
            BLOCKS[i]=BLOCKS[i].lower()
        
    
    return ''.join(BLOCKS)
```

```python
# write a file with block size = 4 

OUTPUT = gc_map(SALMONELLA_LIST, 4, .45)

with open('SALMONELLA_gc', 'w') as f:
    f.write(SALMONELLA.split('\n')[0]+ gc_map(SALMONELLA_LIST, 4, .45))
    
OUTPUT = gc_map(SALMONELLA_LIST, 4, .45)
OUTPUT_LIST = []
J = 0
while J< len(OUTPUT)/60:
    OUTPUT_LIST.append(OUTPUT[J:J+60])
    #print(OUTPUT_LIST)
    J += 60
    
"\n".join(OUTPUT_LIST)
```
